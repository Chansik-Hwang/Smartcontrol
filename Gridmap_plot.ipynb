```python
from google.colab import drive
drive.mount('/content/gdrive')  # default path : /content/gdrive/MyDrive
from matplotlib import colors
import numpy as np
import matplotlib.pyplot as plt
```

```python
def find_row_col(contents):
  row,col=0,0
  for content in contents:
    if(content == ' '):
      col+=1
    elif(content == '\n'):
      break
  for content in contents:
    if(content == '\n'):
      row+=1
  col+=1
  row+=1

  return row,col
```

```python
file_path = '/content/gdrive/MyDrive/Astar_path_Final.txt'
result_file = open(file_path, 'r')
content = result_file.read()
#print(content)
```

```python
grid_list = []
row,col = find_row_col(content)
print(row)
print(col)
for number in content:
  if number == ' ' or number == '\n':
    continue
  elif number == '*':
    grid_list.append(int(5))
  else:
    grid_list.append(int(number))

#print(grid_list)
```

```python
GridArr = np.array(grid_list).reshape(row,col)
#print(GridArr)
col_stepsize = int(col/10) # 실제 map에서 10 값 조정하기
row_stepsize = int(row/10) # 실제 map에서 10 값 조정하기

colormap = colors.ListedColormap(["white","dimgray","lime","red", "gold"])
plt.figure(figsize=(5,5))
title_font = {'fontsize' : 13, 'fontweight' : 'normal'} #글자 크기 필요 시 조정
plt.title("Result path by A star",fontdict=title_font)
plt.xticks(size = 10, color = "black",ticks=np.arange(0,col,col_stepsize))
plt.yticks(size = 10, color = "black",ticks=np.arange(0,row,row_stepsize))
plt.imshow(GridArr, cmap=colormap, )
plt.savefig('/content/gdrive/MyDrive/Gridmap_plot.png', facecolor='#eeeeee', bbox_inches='tight')
```
